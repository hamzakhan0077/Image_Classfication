```python
import os
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing import image_dataset_from_directory
```

```python
best_network = load_model("cifar_transfer_model.h5")
```

```python
path = "examples"
pathnames = [os.path.join(path, filename) for filename in sorted(os.listdir(path))]

imgs = [load_img(img_path, target_size=(224, 224)) for img_path in pathnames]
for img in imgs:
    plt.figure()
    plt.imshow(img)
```

<p>Choosing eight new images from google as the test set as already been used during evaluation</p>
<p> Network has most likely not seen those images before as I  berry picked mercedes, lamborghinig boats, and British airways planes</p>
<p>The reason that I  did this to see how accurate is the network, Does it actuall  thinks of car as a car or ?
    Lets Find out </p>

```python
demo_dataset = image_dataset_from_directory(
    directory="examples", shuffle=False, label_mode=None, image_size=(224, 224), )
```

```python
best_network.predict(demo_dataset) # alphabetic -- class 1 is boats , class 2 - cars, class 3 -planes
# rows are images -- image 1 , images 2 etc etc
```

<p> Lets check these Images one by one to evaluate the strength of our best network.
    
    The line of code above outputs an Array.
    This Array consists of 8 Arrays(because 8 images) with 3
    elements in each array corresponding to our Three Classes.
    those are alphatically ordered I.E the First, Second and Third elements 
    in the are sub array are Boats, Cars , Planes (B,C,P - ALphabetical)
    
    Those elements are probablities of how likely this image is either 
    a Boat, a Car or an Airplane.
    
    For example we can see as a Human that Image 1 and Image 2 is an Airplane
    in the Above output we can see that network gives it 1.0 Probablity 
    for the third element of Sub Array 1 and 2. 
    
    this means that the netweork is 100 % sure that the image is of an Airplane
    however it also think on 6% and 5% probablity that it could be a boat.
    
    Same reasoning can be followed for all the other images.
    
    This is the strength of  my Best Network this the reason I picked it as it is accurate.
    
    However there is a downside to it
    
    consider image 4 it as Mercedes E class, I as a Human can clearly see that.
    However my model is 100% sure that it is a car but it is also wonders if it could be a boat
     [1.42144448e-29, 1.00000000e+00, 0.00000000e+00]

     the reason  That I think that it get the things right is because the weights  that we used during the training 
    were pretrained on CIFAR-10 dataset (Link given below). This dataset has imaghes of all of the 3 classes used during the
    traning along with range of other classes. Since it was pretraind and was then trained on my dataset it has a considerably high accuracy.
